```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
df = spark.read.text('data/word_count.text')
```

```python
words = df.rdd.flatMap(lambda line: line[0].split(' '))
```

```python
count = words.countByValue()
```

```python
for word, count in count.items():
    print('{} : {}'.format(word, count))
```
